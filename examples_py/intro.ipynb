# Introduction to the Python binding of DelFEM2  


## Install
To install the DelFEM2 python module, first donwload the DelFEM2 codes from the GitHub repository
```
git clone https://github.com/nobuyuki83/delfem2
```

Then download and compile the submodule 
```
git submodule init 
git submodule update
cd external/glfw
cmake .
make
```

Compile the python module with
```
cd python
mkdir buildMake
cd buildMake
cmake ..
make
```

If everything goes well, you will see `dfm2_.so` in the module directory `python/dfm2` and you are ready to go~!

## Importing the Module

To use the module, first you need to set the python path to this module. 

````python
import sys
sys.path.append("../module_py") # adding the python path to the dfm2 module
import dfm2
````

## Importing 3D Mesh

The Delfem2 has a mesh class `MeshElem`. The `MeshElem` support 2D and 3D meshe with a single element type. 

````python
mshelm = dfm2.MeshElem("../test_inputs/bunny_2k.ply")
print("length of the element list",len(mshelm.listElem)) # list of element
print("length of the coordinate list",len(mshelm.listPos)) # list of positions
print(mshelm.elemType) # element type
print(mshelm.nDim) # dimension of the coordinate
````

## Draw 3D Mesh

you can draw a 3D mesh using image viewer or 3D window. In this Jupyter notebook page, we only demonstrate Drawing the mesh using the image viewer such as the `pyplot`.

The function 'imgDraw3d' outputs byte RGB array of the image. We can directory draw the image with the `pyplot` as

````python
%matplotlib inline
import matplotlib.pyplot as plt

mshelm = dfm2.MeshElem("../test_inputs/bunny_2k.ply")
mshelm.scaleXYZ(0.03)
# dfm2.winDraw3D(mshelm,winsize=(400,300))  #Open new window to draw the mesh
img = dfm2.imgDraw3d(mshelm,winsize=(400,300))
plt.imshow(img)
plt.show()
````
